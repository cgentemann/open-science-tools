```python
#write a function to take a dataframe, iterate over both score columns, and return 
#the percentages of papers that have each score
def calc_percents(df):
    '''
    This function takes the individual evaluation spreadsheetsand calculates the percentage of papers that scored compliant
    '''
    df = df.reset_index()
    software_ntwo = 0
    software_none = 0
    software_zero = 0
    software_one = 0
    data_ntwo = 0
    data_none = 0
    data_zero = 0
    data_one = 0
    for i in range(30):
        if df["Software availability score"][i]==-2:
            software_ntwo += 1
        elif df["Software availability score"][i]==-1:
            software_none += 1
        elif df["Software availability score"][i]==0:
            software_zero += 1
        elif df["Software availability score"][i]==1:
            software_one += 1
        else:
            print("Unknown score encountered at row {x}".format(x=i))
    for i in range(30):
        if df["Data accessibility score"][i]==-2:
            data_ntwo += 1
        elif df["Data accessibility score"][i]==-1:
            data_none += 1
        elif df["Data accessibility score"][i]==0:
            data_zero += 1
        elif df["Data accessibility score"][i]==1:
            data_one += 1
        else:
            print("Unknown score encountered at row {x}".format(x=i))
    data_perc = (data_zero+data_one)/30 *100
    soft_perc = (software_zero+software_one)/30 * 100
    return round(data_perc,2), round(soft_perc,2)
    
```

```python
#create a table with compliance perccentages
d = {"Division": ["APD","ESD","HPD","PSD","BPS"], 
     "% Compliant (Data)": [calc_percents(apd)[0], calc_percents(esd)[0],calc_percents(hpd)[0] , calc_percents(psd)[0],calc_percents(bps)[0]],
    "% Compliant (Software)": [calc_percents(apd)[1], calc_percents(esd)[1],calc_percents(hpd)[1] , calc_percents(psd)[1],calc_percents(bps)[1]]}
compliance_table = pd.DataFrame(d)
```

# Estimate of Errors

Esimate the errors associated with the measurements.  The errors will likely be an underestimate due to other factors that could contribute to the measurement, but they provide some estimate for the uncertainty. 

```python
def pop_error(percent, sample, Z=1):
    """
    Calculate the error from sampling an population
   
    percent: Percentage with a certain classifictaion
    sample: Number of samples taken
    Z: z-value for level of confidence
   
    Source: https://en.wikipedia.org/wiki/Population_proportion
    """
    n_o = sample
   
    return Z * (percent * ( 1 - percent) / n_o)**0.5
```

```python
def error_sampling_estimate(percent, sample, total, niter=100):
    """Determing a samping error given a percent, total sample, and number of samples
   
    percent: Percentage with a certain classifictaion
    sample: Number of samples taken
    total: Total number of samples
    niter: Number of times for iteration
   
    """
    repro_arr = (np.random.uniform(0,1,total) < percent)
    x = np.zeros(niter)
    for j in range(0,niter):
        rid = np.random.randint(0,total, sample)
        x[j] = repro_arr[rid].sum()/sample
    return x.mean(), x.mean() - percent, x.std()
```

```python
COL_NOT   = "#d73027"  # red (bad) - TRANSPARENT
COL_PART  = "#66c2a5"  # teal green (partial) - SOLID
COL_FULL  = "#1a9850"  # dark green (good+) - SOLID
ALPHA_NOT = 0.2
ALPHA_PART = 1.0
ALPHA_FULL = 1.0
```

```python
# Helper for readable text
def text_color(hex_color, alpha=1.0):
    hex_color = hex_color.lstrip('#')
    r, g, b = [int(hex_color[i:i+2], 16) for i in (0, 2, 4)]
    luminance = 0.2126*r + 0.7152*g + 0.0722*b
    # For transparent colors, use black text
    if alpha < 0.5:
        return 'black'
    return 'black' if luminance > 140 else 'white'
```

```python

plt.figure(figsize=(9, 5))

bar_h = 0.6
x = np.arange(len(dd))

nc = dd['Not Compliant'][::-1].to_numpy() / 60 * 100 
pc = dd['Partially Compliant'][::-1].to_numpy() / 60 * 100
fc = dd['Fully Compliant'][::-1].to_numpy() / 60  * 100

plt.barh(x, nc, color=COL_NOT, height=bar_h, alpha=ALPHA_NOT)
plt.barh(x, pc, left=nc, color=COL_PART, height=bar_h, alpha=ALPHA_PART)
plt.barh(x, fc, left=nc+pc, color=COL_FULL, height=bar_h, alpha=ALPHA_FULL)


# Percentage labels in every segment

for i in range(len(dd)):
    start = 0
    
    for w, c, a in zip([nc[i], pc[i], fc[i]],
                        [COL_NOT, COL_PART, COL_FULL],
                        [ALPHA_NOT, ALPHA_PART, ALPHA_FULL]):
            if w >= 4:
                plt.text(start + w/2, i, f"{w:.0f}%", ha='center', va='center',
                         color=text_color(c, a), fontsize=10, fontweight='bold')
            start += w

# Category labels positioned above the chart
avg_not_width = np.mean(nc)
avg_part_width = np.mean(pc)
avg_full_width = np.mean(fc)

label_y = len(dd) - 0.4
    
plt.text(10, label_y, 'Not Compliant', 
             ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')
plt.text(avg_not_width + avg_part_width/2, label_y, 'Partially Compliant', 
             ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')
plt.text(37 + avg_part_width + avg_full_width/2, label_y, 'Fully Compliant', 
             ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')

# Axes and formatting
plt.yticks(x, dd['Division'][::-1], fontsize=12)
plt.xlabel('Papers with Data Available (%)', fontsize=14)
plt.title(f'Data Compliance Status by Division', fontsize=16, pad=20)
plt.xlim(0, 100)
plt.ylim(-0.5, len(dd) + 0.2)
plt.grid(axis='x', linestyle='--', alpha=0.35)
plt.tight_layout()


plt.savefig('data_compliance_bar.jpg', dpi=300, bbox_inches='tight')

plt.show()
```
